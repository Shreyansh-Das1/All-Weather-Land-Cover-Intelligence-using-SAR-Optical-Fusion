```python
import numpy as np
import os
```

```python
labels_path = r"D:\Resume Projs\All-Weather-Land-Cover-Intelligence-using-SAR-Optical-Fusion\data\dataset\labels"

classcount= {0:0, 1:0, 2:0, 3:0, 4:0}
tiles = os.listdir(labels_path)
```

```python
for file in tiles:
    tilepath = os.path.join(labels_path, file)
    tile = np.load(tilepath)
    clas , count = np.unique(tile, return_counts= True)
for c,cnt in zip(clas, count):
    classcount[c] += cnt
```

```python
classSum =0 
for classes in classcount:
    print(classcount[classes])
    classSum += classcount[classes]
```

```python
classwt =[0,0,0,0,0]
for i in range(1,5):
    classwt[i] = classSum/(5*classcount[i])
print(classwt)
```
